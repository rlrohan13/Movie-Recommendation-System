#### Column format correction

Now lets correcting the column format for columns such as genres, keywords, cast, and crew.

##### genres col

```python
movies.iloc[0].genres
```

We want to convert this dictionary [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}] into the formate:

['Action','Adventure','Fantasy','Scifi'].

```python
# creating a fn:

def convert(obj):
  L = []                      # L is empty list where we will append (add) the name key from genres
  for i in obj:
    L.append(i['name'])       # appending the name key from genres
  return L
```

```python
convert([{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}])
```

But some times we might get error like string indices must be integer.

Because each element in this list: [{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}] is a dictionary and therefore each indices is a string.

So use a module 'ast' and there is one fn in it called ast.literal_eval. This will convert the string indices into intergers.

```python
# example:

import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

Therefore using the ast module:

```python
import ast

def convert(text):
    if isinstance(text, str):  # Check if the input is a string
        L = []
        for i in ast.literal_eval(text):        #  ast.literal_eval will convert string indices into integers
            L.append(i['name'])
        return L
    else:
        return text  # Return the input as is if it's not a string
```

```python
movies['genres'] = movies['genres'].apply(convert)  # applying the convert fn to genres col
movies['genres'].head(5)
```

##### cast col

Here, we want the initial 3 dictionaies since we want only the top 3 actors.

```python
movies['cast'][0]
```

So we want the name of the initial 3 actors i.e., the name key from the initial 3 dictionaries:

{"cast_id": 242, "character": "Jake Sully", "credit_id": "5602a8a7c3a3685532001c9a", "gender": 2, "id": 65731, "name": "Sam Worthington", "order": 0} : *"name": "Sam Worthington"*

{"cast_id": 3, "character": "Neytiri", "credit_id": "52fe48009251416c750ac9cb", "gender": 1, "id": 8691, "name": "Zoe Saldana", "order": 1}, *"name": "Zoe Saldana"*

{"cast_id": 25, "character": "Dr. Grace Augustine", "credit_id": "52fe48009251416c750aca39", "gender": 1, "id": 10205, "name": "Sigourney Weaver", "order": 2} : *"name": "Sigourney Weaver"*

```python
def convert3(text):
    L = []                    # L = empty list where we will store the values of the initial name key (means storing the name of top 3 stars in the movies)
    counter = 0                              # counter value will start from 0.
    for i in ast.literal_eval(text):
        if counter != 3:                         # means counter value (indicates the indices value) will start from 0 and go till 2 i.e., 0,1,2. Therefore we will get the inital 3 dictionaries.
          L.append(i['name'])                 # we will add (append) the value of the name key in the empty list L.
          counter+=1                           # counter value will increase by 1 if condition holds true.
        else:
          break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
movies['cast'].head()
```

##### crew col

We want only a director name for a movie in the crew col.

```python
movies['crew'].head(1)
```

We can see that just for 1 row (one movie), there is a lot of crew members.

And 'job' is the key which tells about what work a particular crew member has done.

Therefore, we need to search the 'job' key for each crew in order to get the name of the director.

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break                     # means the moment the it got the director name, the loop will break and will not go to the other dictionaries and the loop will start again for next row (next movie)
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

#### Removing space

eg: we want:

'Sam Worthington' = 'SamWorthington'

'Sam Mendes' = 'SamMendes'

we don't want any space btw two items in a single elements to avoid confusion during recommendation because 'Sam Worthington' and 'Sam Mendes' are two different person but due to same name, there might be some confusion.

Therefore, the same thing we need to do for every element which contains 2 items: eg: 'culture clash' to 'cultureclash'.

```python
# creating a fn for removing the space:

def collapse(L):
    L1 = []                # L1 = empty list
    for i in L:
        L1.append(i.replace(" ",""))      # it will replace the empty space by no space.
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

## Saming

it is used to make the similar words having same meaning as one word.

Eg = actions and action are one word.

     loving, loved, love are one word.

```python
! pip install nltk

import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)          # it will return a string
```

```python
# eg:

['loved', 'loving', 'love']
['love', 'love', 'love']
```

## Recommendation fn

we have to make a fn where if we ask a movie, then it will recommend me the 5 most similar movies.

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6]          # it will sort the distances in descending order so that we can

# [1:6] means we want the initial 5 most similar movie

    for i in movies_list:
      print(i[0])
```

```python
recommend('Avatar')
```
